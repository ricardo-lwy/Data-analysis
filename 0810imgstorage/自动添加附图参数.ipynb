```python
####库
from pandas.core.frame import DataFrame
from email.mime import image
from openpyxl import load_workbook
import pandas as pd
from openpyxl.writer.excel import ExcelWriter
import re


####辅助函数


###1.字符自动转化   特殊符号替换同时大写转小写
def transtrtype(mystr):
    translationTable = str.maketrans("éàèùâêîôûç'_", "eaeuaeiouc--")
    newstr=mystr.translate(translationTable).lower()
    return newstr
####str.lower()# 把所有字符中的大写字母转换成小写字母
##########
####08/10/2022 匹配成功


###2.插入一条数据在表中

def insert(df, i, df_add):
        # 指定第i行插入一行数据
    df1 = df.iloc[:i, :]
    df2 = df.iloc[i:, :]
    df_new = pd.concat([df1, df_add, df2], ignore_index=True)
    return df_new
```
